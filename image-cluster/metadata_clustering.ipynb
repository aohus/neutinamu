```python
# GPS 데이터 추출 함수
def get_gps_coords(img_path):
    try:
        exif_dict = piexif.load(img_path)
        gps = exif_dict.get("GPS")
        if gps:

            def convert(coord, ref):
                degrees, minutes, seconds = [x[0] / x[1] for x in coord]
                result = degrees + (minutes / 60.0) + (seconds / 3600.0)
                return result if ref in ["N", "E"] else -result

            lat = convert(
                gps[piexif.GPSIFD.GPSLatitude],
                gps[piexif.GPSIFD.GPSLatitudeRef].decode(),
            )
            lon = convert(
                gps[piexif.GPSIFD.GPSLongitude],
                gps[piexif.GPSIFD.GPSLongitudeRef].decode(),
            )
            return np.array([lat, lon])
    except:
        pass
    return np.array([0, 0])
```
